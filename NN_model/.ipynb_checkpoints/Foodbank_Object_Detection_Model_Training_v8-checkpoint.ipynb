# Create dataloader

```python
class CustomTensorDataset(Dataset):
    # initialize the constructor
    def __init__(self, tensors, transforms=None):
      self.tensors = tensors
      self.transforms = transforms

    def __getitem__(self, index):
        # grab the image, label, and its bounding box coordinates
        image = self.tensors[0][index]
        label_prdtype = self.tensors[1][index]
        label_weight = self.tensors[2][index]
        label_halal = self.tensors[3][index]
        label_healthy = self.tensors[4][index]
        bbox = self.tensors[5][index]
        filename = self.tensors[6][index]
        # transpose the image such that its channel dimension becomes
        # the leading one
        image = image.permute(2, 0, 1)
        # check to see if we have any image transformations to apply
        # and if so, apply them
        if self.transforms:
          image = self.transforms(image)
        # return a tuple of the images, labels, and bounding box coordinates
        return (image, label_prdtype, label_weight, label_halal, label_healthy, bbox, filename)

    def __len__(self):
        # return the size of the dataset
        return self.tensors[0].size(0)
```

```python
data = []
labels_prdtype = []
labels_weight = []
labels_halal = []
labels_healthy = []
labels_total = []
bboxes = []
imagePaths = []
filenames = []

# loop over all files
for idx in range(annotations.shape[0]):
    startX = annotations["xmin"][idx]
    startY = annotations["ymin"][idx]
    endX = annotations["xmax"][idx]
    endY = annotations["ymax"][idx]
    filepath = annotations["filepath"][idx]
    filename = annotations["filepath"][idx].rsplit('.', 1)[0]
    label = annotations["label"][idx]
    folder_name = annotations["folder_name"][idx]

    # derive the path to the input image, load the image (in
    # OpenCV format), and grab its dimensions
    imagePath = os.path.join("/content", CONFIGS["DATA_BASE_PATH"], filepath)
    image = cv2.imread(imagePath)
    (h, w) = image.shape[:2]
    # scale the bounding box coordinates relative to the spatial
    # dimensions of the input image
    startX = float(startX) / w
    startY = float(startY) / h
    endX = float(endX) / w
    endY = float(endY) / h
    # load the image and preprocess it
    # image = cv2.imread(imagePath)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    image = cv2.resize(image, (224, 224))
    data.append(image)
    bboxes.append((startX, startY, endX, endY))
    imagePaths.append(imagePath)
    filenames.append(filename)

labels_prdtype = annotations['ProductType']
labels_weight = annotations['Weight']
labels_halal = annotations['HalalStatus']
labels_healthy = annotations['HealthStatus']
```

# Train the neural network via transfer learning

```python
from torchvision.models import resnet101

# define normalization and data augmentation transforms
transforms_imgs = transforms.Compose([
    transforms.ToPILImage(),
    transforms.RandomHorizontalFlip(p=0.5),  # randomly flip images horizontally
    transforms.RandomVerticalFlip(p=0.5),  # randomly flip images vertically
    transforms.RandomRotation(20),  # randomly rotate images by 20 degrees
    transforms.ColorJitter(brightness=0.1, contrast=0.1, saturation=0.1, hue=0.1),  # randomly change color properties
    transforms.ToTensor(),
    transforms.Normalize(mean=CONFIGS['IMG_MEAN'], std=CONFIGS['IMG_STD'])
])


# convert NumPy arrays to PyTorch datasets
trainDS = CustomTensorDataset((trainImages, trainLabels_prdtype, trainLabels_weight, trainLabels_halal, trainLabels_healthy,
                               trainBBoxes, trainFilenames), transforms=transforms_imgs)
testDS = CustomTensorDataset((testImages, testLabels_prdtype, testLabels_weight, testLabels_halal, testLabels_healthy,
                              testBBoxes, testFilenames), transforms=transforms_imgs)
print("[INFO] total training samples: {}...".format(len(trainDS)))
print("[INFO] total test samples: {}...".format(len(testDS)))
# calculate steps per epoch for training and validation set
trainSteps = len(trainDS) // CONFIGS['BATCH_SIZE']
valSteps = len(testDS) // CONFIGS['BATCH_SIZE']

# create data loaders
trainLoader = DataLoader(trainDS, batch_size=CONFIGS['BATCH_SIZE'],
	shuffle=True, num_workers=os.cpu_count(), pin_memory=CONFIGS['PIN_MEMORY'])
testLoader = DataLoader(testDS, batch_size=CONFIGS['BATCH_SIZE'],
	num_workers=os.cpu_count(), pin_memory=CONFIGS['PIN_MEMORY'])
```

```python
from torchvision import models
import torch.nn as nn
from torch.nn import CrossEntropyLoss

# Model Architecture
base_model = models.resnet50(pretrained=True)
num_features = base_model.fc.in_features
base_model.fc = nn.Identity()

class CustomResNet(nn.Module):
    def __init__(self, base_model, num_categories1, num_categories2, num_categories3, num_categories4):
        super(CustomResNet, self).__init__()
        self.base_model = base_model
        self.hidden_dim = 512  # Define hidden layer dimensions

        # Head for ProductType
        self.fc1_1 = nn.Linear(num_features, self.hidden_dim)
        self.fc1_2 = nn.Linear(self.hidden_dim, self.hidden_dim)
        self.fc1_out = nn.Linear(self.hidden_dim, num_categories1)

        # Head for Weight
        self.fc2_1 = nn.Linear(num_features, self.hidden_dim)
        self.fc2_2 = nn.Linear(self.hidden_dim, self.hidden_dim)
        self.fc2_out = nn.Linear(self.hidden_dim, num_categories2)

        # Head for HalalStatus
        self.fc3_1 = nn.Linear(num_features, self.hidden_dim)
        self.fc3_2 = nn.Linear(self.hidden_dim, self.hidden_dim)
        self.fc3_out = nn.Linear(self.hidden_dim, num_categories3)

        # Head for HealthStatus
        self.fc4_1 = nn.Linear(num_features, self.hidden_dim)
        self.fc4_2 = nn.Linear(self.hidden_dim, self.hidden_dim)
        self.fc4_out = nn.Linear(self.hidden_dim, num_categories4)

        # Head for BoundingBox
        self.fc_bbox = nn.Linear(num_features, 4)

    def forward(self, x):
        x = self.base_model(x)

        # Head for ProductType
        x1 = nn.functional.relu(self.fc1_1(x))
        x1 = nn.functional.relu(self.fc1_2(x1))
        out1 = self.fc1_out(x1)

        # Head for Weight
        x2 = nn.functional.relu(self.fc2_1(x))
        x2 = nn.functional.relu(self.fc2_2(x2))
        out2 = self.fc2_out(x2)

        # Head for HalalStatus
        x3 = nn.functional.relu(self.fc3_1(x))
        x3 = nn.functional.relu(self.fc3_2(x3))
        out3 = self.fc3_out(x3)

        # Head for HealthStatus
        x4 = nn.functional.relu(self.fc4_1(x))
        x4 = nn.functional.relu(self.fc4_2(x4))
        out4 = self.fc4_out(x4)

        # Head for BoundingBox
        bbox = self.fc_bbox(x)

        return out1, out2, out3, out4, bbox


# Initialize the CustomResNet model
base_model = resnet50(pretrained=True)
# Remove the final fully connected layer (classifier)
base_model.fc = nn.Identity()
custom_resnet_model = CustomResNet(base_model, len(le_prdtype.classes_), len(le_weight.classes_), len(le_halal.classes_), len(le_healthy.classes_))
custom_resnet_model = custom_resnet_model.to(CONFIGS['DEVICE'])

# Initialize loss functions and optimizer
classLossFunc = CrossEntropyLoss()
bboxLossFunc = MSELoss()
opt = Adam(custom_resnet_model.parameters(), lr=CONFIGS['INIT_LR'], weight_decay=0.0001)

# Initialize history and other variables
H = {
    "total_train_loss": [], "total_val_loss": [],
    "train_class_total_acc": [], "val_class_total_acc": [],
    "train_class_prdtype_acc": [], "val_class_prdtype_acc": [],
    "train_class_weight_acc": [], "val_class_weight_acc": [],
    "train_class_halal_acc": [], "val_class_halal_acc": [],
    "train_class_healthy_acc": [], "val_class_healthy_acc": []
}

# Training loop
for e in range(CONFIGS['NUM_EPOCHS']):
    custom_resnet_model.train()
    totalTrainLoss = 0
    trainCorrect_total = 0
    trainCorrect_prdtype = 0
    trainCorrect_weight = 0
    trainCorrect_halal = 0
    trainCorrect_healthy = 0

    for (images, labels_prdtype, labels_weight, labels_halal, labels_healthy, bboxes, filenames) in trainLoader:
        images, bboxes = images.to(CONFIGS['DEVICE']), bboxes.to(CONFIGS['DEVICE'])

        # Forward pass
        out1, out2, out3, out4, pred_bboxes = custom_resnet_model(images)
        # Compute individual losses
        loss1 = classLossFunc(out1, labels_prdtype.to(CONFIGS['DEVICE']))
        loss2 = classLossFunc(out2, labels_weight.to(CONFIGS['DEVICE']))
        loss3 = classLossFunc(out3, labels_halal.to(CONFIGS['DEVICE']))
        loss4 = classLossFunc(out4, labels_healthy.to(CONFIGS['DEVICE']))
        bboxLoss = bboxLossFunc(pred_bboxes, bboxes)

        # Compute loss
        totalLoss = CONFIGS['BBOX'] * bboxLoss + \
            CONFIGS['LABELS_PRDTYPE'] * loss1 + \
            CONFIGS['LABELS_WEIGHT'] * loss2 + \
            CONFIGS['LABELS_HALAL'] * loss3 + \
            CONFIGS['LABELS_HEALTHY'] * loss4

        # Backward pass and optimization
        opt.zero_grad()
        totalLoss.backward()
        opt.step()

        # Update training loss and accuracy for each head
        totalTrainLoss += totalLoss.item()
        trainCorrect_prdtype += (out1.argmax(1) == labels_prdtype.to(CONFIGS['DEVICE'])).float().sum().item()
        trainCorrect_weight += (out2.argmax(1) == labels_weight.to(CONFIGS['DEVICE'])).float().sum().item()
        trainCorrect_halal += (out3.argmax(1) == labels_halal.to(CONFIGS['DEVICE'])).float().sum().item()
        trainCorrect_healthy += (out4.argmax(1) == labels_healthy.to(CONFIGS['DEVICE'])).float().sum().item()
        trainCorrect_total += ((out1.argmax(1) == labels_prdtype.to(CONFIGS['DEVICE'])) & (out2.argmax(1) == labels_weight.to(CONFIGS['DEVICE'])) &
                              (out3.argmax(1) == labels_halal.to(CONFIGS['DEVICE'])) & (out4.argmax(1) == labels_healthy.to(CONFIGS['DEVICE']))).float().sum().item()


    # Validation loop
    custom_resnet_model.eval()
    totalValLoss = 0
    valCorrect_total = 0
    valCorrect_prdtype = 0
    valCorrect_weight = 0
    valCorrect_halal = 0
    valCorrect_healthy = 0

    with torch.no_grad():
        for (images, labels_prdtype, labels_weight, labels_halal, labels_healthy, bboxes, filenames) in testLoader:
            images, bboxes = images.to(CONFIGS['DEVICE']), bboxes.to(CONFIGS['DEVICE'])

            # Forward pass
            out1, out2, out3, out4, pred_bboxes = custom_resnet_model(images)

            # Compute loss
            loss1 = classLossFunc(out1, labels_prdtype.to(CONFIGS['DEVICE']))
            loss2 = classLossFunc(out2, labels_weight.to(CONFIGS['DEVICE']))
            loss3 = classLossFunc(out3, labels_halal.to(CONFIGS['DEVICE']))
            loss4 = classLossFunc(out4, labels_healthy.to(CONFIGS['DEVICE']))
            bboxLoss = bboxLossFunc(pred_bboxes, bboxes)
            totalLoss = CONFIGS['BBOX'] * bboxLoss + \
                CONFIGS['LABELS_PRDTYPE'] * loss1 + \
                CONFIGS['LABELS_WEIGHT'] * loss2 + \
                CONFIGS['LABELS_HALAL'] * loss3 + \
                CONFIGS['LABELS_HEALTHY'] * loss4

            # Update validation loss and accuracy
            totalValLoss += totalLoss.item()
            valCorrect_prdtype += (out1.argmax(1) == labels_prdtype.to(CONFIGS['DEVICE'])).float().sum().item()
            valCorrect_weight += (out2.argmax(1) == labels_weight.to(CONFIGS['DEVICE'])).float().sum().item()
            valCorrect_halal += (out3.argmax(1) == labels_halal.to(CONFIGS['DEVICE'])).float().sum().item()
            valCorrect_healthy += (out4.argmax(1) == labels_healthy.to(CONFIGS['DEVICE'])).float().sum().item()
            valCorrect_total += ((out1.argmax(1) == labels_prdtype.to(CONFIGS['DEVICE'])) & (out2.argmax(1) == labels_weight.to(CONFIGS['DEVICE'])) &
                                (out3.argmax(1) == labels_halal.to(CONFIGS['DEVICE'])) & (out4.argmax(1) == labels_healthy.to(CONFIGS['DEVICE']))).float().sum().item()


    # Calculate and store average losses and accuracies
    avgTrainLoss = totalTrainLoss / len(trainLoader)
    avgValLoss = totalValLoss / len(testLoader)

    trainAcc_prdtype = trainCorrect_prdtype / len(trainDS)
    trainAcc_weight = trainCorrect_weight / len(trainDS)
    trainAcc_halal = trainCorrect_halal / len(trainDS)
    trainAcc_healthy = trainCorrect_healthy / len(trainDS)
    trainAcc_total = trainCorrect_total / len(trainDS)

    valAcc_prdtype = valCorrect_prdtype / len(testDS)
    valAcc_weight = valCorrect_weight / len(testDS)
    valAcc_halal = valCorrect_halal / len(testDS)
    valAcc_healthy = valCorrect_healthy / len(testDS)
    valAcc_total = valCorrect_total / len(testDS)

    # Update history
    H["total_train_loss"].append(avgTrainLoss)
    H["total_val_loss"].append(avgValLoss)
    H["train_class_prdtype_acc"].append(trainAcc_prdtype)
    H["train_class_weight_acc"].append(trainAcc_weight)
    H["train_class_halal_acc"].append(trainAcc_halal)
    H["train_class_healthy_acc"].append(trainAcc_healthy)
    H["train_class_total_acc"].append(trainAcc_total)

    H["val_class_prdtype_acc"].append(valAcc_prdtype)
    H["val_class_weight_acc"].append(valAcc_weight)
    H["val_class_halal_acc"].append(valAcc_halal)
    H["val_class_healthy_acc"].append(valAcc_healthy)
    H["val_class_total_acc"].append(valAcc_total)

    # Print epoch summary
    print(f"EPOCH {e+1}/{CONFIGS['NUM_EPOCHS']}")
    print(f"Train Loss: {avgTrainLoss}, Val Loss: {avgValLoss}")
    print(f"Train Acc - Prdtype: {trainAcc_prdtype}, Weight: {trainAcc_weight}, Halal: {trainAcc_halal}, Healthy: {trainAcc_healthy}, Total: {trainAcc_total}")
    print(f"Val Acc - Prdtype: {valAcc_prdtype}, Weight: {valAcc_weight}, Halal: {valAcc_halal}, Healthy: {valAcc_healthy}, Total: {valAcc_total}")

```
